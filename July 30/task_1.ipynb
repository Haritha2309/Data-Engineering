CREATE A CSV OF SALES DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000

Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""
with open('sales_data.csv', 'w') as file:
  file.write(csv_data)

df = pd.read_csv('sales_data.csv')
print(df)
```

TOTAL SALES PER STORE

```python
grouped = df.groupby('Store')['Sales'].sum().reset_index()
print("TOTAL SALES PER STORE : \n",grouped)
```

MONTHLY SALES TREND

```python
month_grp = df.groupby('Month')['Sales'].sum().reset_index()
print("MONTHLY SALES TREND : \n", month_grp)
```

LINE CHART

```python
import matplotlib.pyplot as plt
import pandas as pd

month_order = ['Jan','Feb','Mar']
df['Month']= pd.Categorical(df['Month'],categories=month_order)
pivot_df = df.pivot(index = 'Month',columns='Store',values='Sales')
pivot_df.plot(kind = 'line',marker = '*',color = ['black','red'],figsize=(5,3))
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.legend(loc = 'lower right',title = 'STORE')
plt.grid(True)
plt.show()
```

BAR CHART

```python
import matplotlib.pyplot as plt
import pandas as pd

store_sales = df.groupby('Store')['Sales'].sum().reset_index()
#print(store_sales)
plt.figure(figsize=(4,3))
plt.xlabel('STORES')
plt.ylabel('SALES')
plt.title('SALES CHART')
bars = plt.bar(store_sales['Store'],store_sales['Sales'],color = ['black','orange'],width=0.5)
for bar in bars:
  height = bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,height, f'{height}', ha='center', va='bottom', fontsize=6)
plt.savefig('sales_chart.png',dpi=300)
plt.show()
```
